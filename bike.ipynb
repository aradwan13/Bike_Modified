```python
from keras.models import Sequential
from keras.layers import Dense,ReLU,BatchNormalization,LeakyReLU,Dropout
```

```python
ann = Sequential()
```

```python
ann.add(Dense(units=25,input_dim=10))
```

```python
ann.add(ReLU())
```

```python
ann.add(Dropout(0.2))
```

```python
ann.add(Dense(units=25))
```

```python
ann.add(ReLU())
```

```python
ann.add(Dropout(0.5))
```

```python
#ann.add(BatchNormalization())
```

```python
ann.add(Dense(units=1, activation = 'linear'))
```

```python
ann.summary()
```

```python
ann.compile(optimizer='adam',loss='mse')
```

```python
ann.fit(X_train,y_train,batch_size=20,epochs=50)
```

```python
pd.DataFrame(ann.history.history)
```

```python
pd.DataFrame(ann.history.history).plot()
```

```python
y_pred = ann.predict(X_test)
```
